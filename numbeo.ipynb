# Making your next move using [Numbeo](https://www.numbeo.com)

```python
import pandas as pd

current_city = "Stockholm, Sweden"
```

## Purchasing Power

```python
purchasing_power = pd.read_html("https://www.numbeo.com/cost-of-living/rankings_current.jsp")[2]
purchasing_power_column = "Local Purchasing Power Index"
purchasing_power = purchasing_power[["City", purchasing_power_column]]


purchasing_power_current = purchasing_power[purchasing_power.City == current_city].iloc[0]

purchasing_power = purchasing_power[purchasing_power[purchasing_power_column] > purchasing_power_current[purchasing_power_column]]
purchasing_power.sort_values(purchasing_power_column, ascending = False)
```

## Property Prices

```python
property_prices = pd.read_html("https://www.numbeo.com/property-investment/rankings_current.jsp")[2]
price_to_income_ratio_column = "Price To Income Ratio"
property_prices = property_prices[["City", price_to_income_ratio_column]]

property_prices_current = property_prices[property_prices.City == current_city].iloc[0]

property_prices = property_prices[property_prices[price_to_income_ratio_column] < property_prices_current[price_to_income_ratio_column]]
property_prices.sort_values(price_to_income_ratio_column)
```

```python
#df_combined = purchasing_power.set_index("City").join(property_prices.set_index("City"), how = "inner").reset_index()

df_combined = pd.merge(purchasing_power, property_prices, on = "City")
df_combined["Country"] = df_combined.City.str.split(", ").str[-1]

countries_interested = ["United States", "Canada", "Australia", "United Kingdom", "Ireland", "South Africa", "New Zealand"]
df_combined = df_combined[df_combined.Country.isin(countries_interested)]

df_combined.sort_values("Local Purchasing Power Index", ascending = True)
```
